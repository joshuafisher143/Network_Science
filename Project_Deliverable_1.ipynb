### Dataset & Business Use Case

In the late 1990’s, the NHL started their initiative to track player and team statistics on a game-by-game basis. However, it was not until this past decade that sports analytics came into the forefront as a tool to measure player performance. Teams around the league have established analytics teams to track and model a multitude of stats in order evaluate what make a player good and therefore what wins hockey games. For this project, I will be acting as the general manager for the upcoming NHL expansion team, the Seattle Kraken. With the expansion draft coming up, I want to ready myself by identifying who the key players are and what are the metrics that I can use to identify them. I will be using a collection of NHL datasets that capture a season’s worth of game and player statistics. The repository can be found at:

https://www.kaggle.com/martinellis/nhl-game-data. Specifically, I will be using the datasets entitled: “game”, “player_info”, “game_skater_stats”, “game_plays” and “game_plays_players”. Some pre-processing will be required to get the data in the right format. I will merge these datasets and filter out unnecessary columns.

Using the techniques and tools that we learn in class, I aim to answer the question: “Which players around the league offer the most value to their teams and what are the metrics that support these findings?” My model will be divided into three core components: games, game stats, and play-by-play stats. This model design will allow me to utilize different layers of granularity and thus allow me to fully characterize each player. By identifying these players, I will be able to build a Stanley Cup contending team.
